```python
class SequentialSearch:
    def __init__(self, data):
        self.data = data

    def search(self, target):
        steps = 0
        for index in range(len(self.data)):
            steps += 1
            if self.data[index] == target:
                return index, steps
        return -1, steps


# Testing Sequential Search
arr = [15, 8, 23, 42, 4, 33]
seq = SequentialSearch(arr)

print("Sequential Search Results:")
targets = [15, 23, 33, 4, 99]   # start, middle, end, random, not found

for t in targets:
    index, steps = seq.search(t)
    print(f"Target {t} -> Index: {index}, Steps: {steps}")
```

```python
class BinarySearch:
    def __init__(self, data):
        self.data = sorted(data)

    def search(self, target):
        low = 0
        high = len(self.data) - 1
        steps = 0

        while low <= high:
            steps += 1
            mid = (low + high) // 2

            if self.data[mid] == target:
                return mid, steps
            elif target > self.data[mid]:
                low = mid + 1
            else:
                high = mid - 1

        return -1, steps


# Testing Binary Search
bin_search = BinarySearch(arr)

print("\nBinary Search Results:")
for t in targets:
    index, steps = bin_search.search(t)
    print(f"Target {t} -> Index: {index}, Steps: {steps}")


# ----------------------------------------------
# Task 2C — Why Binary Search is Faster
# ----------------------------------------------
print("\nExplanation: Binary Search cuts the list into half each time,")
print("so the steps grow very slowly (log n),")
print("while Sequential Search checks every element (n).")
```

```python
import random

# create a sorted list (1 to 100,000)
data = list(range(1, 100001))
value = random.choice(data)

print("\nLarge List Experiment:")
print(f"Random Target Selected: {value}")

seq_large = SequentialSearch(data)
bin_large = BinarySearch(data)

idx1, steps1 = seq_large.search(value)
idx2, steps2 = bin_large.search(value)

print(f"Sequential Search Steps: {steps1}")
print(f"Binary Search Steps: {steps2}")
print("Binary Search is dramatically faster on large lists!")
```
